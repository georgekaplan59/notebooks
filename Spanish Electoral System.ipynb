# Spanish Electoral System

## Introduction

This notebook argues that the Spanish electoral system is no longer acceptable. At present, every Spaniard has the formal right to vote for a deputy in Parliament, but their vote does not carry the same value as every other. 

Spanish democracy faces serious problems:

* Bribery and corruption (second main issue for Spaniards).
* Increasing voter discontent with politicians, parties and politics (third main issue for Spaniards).
* Catalan independence (fourth main issue for Spaniards).
* Disproportional influence of pro-independence parties in Parliament.
* Dubious separation of powers.
* Lack of a culture of consensus.

The problems listed above arise from several sources. The electoral system is only one of them. They could be reduced, and hopefully even solved, by the introduction of a greater proportionality. This objective is easily achievable by two modifications in the current electoral law. A more proportional system would likely enhance voter confidence in politics and politicians, encourage popular participation, remove strategic votes, transform legislatures in a system of checks and balances, reduce the weight of nationalism in governability, suppress the differences in the number of votes required per seat among constituencies or districts, and facilitate the making of pre and post-electoral coalitions.

## Electoral systems

According to Farrell (2011), an electoral systems have five main characteristics:

* The _electoral formula_ is the algorithm to convert votes into seats.
* The _ballot structure_: 
    * _categorical_. The voter should choose only one party (closed lists).
    * _ordinal_. The voter should rank some or all of the candidates listed (open lists).
* The _district magnitude_ refers to the number of seats elected from each constituency.
* The _threshold of election_ is the percentage of valid votes required to elect a deputy. 
* _Proportionality_ refers to the accuracy with which the electoral system translates votes into parliamentary representation. It depends of the previous characteristics, so it can be seen as a dependent variable. Lijphart (1990) analyzes the relationship among these variables.

Most Western democracies use some form of proportional system. In this system, the nation is divided into a variable number of constituencies or districts. Each constituency is represented by a number of seats or deputies in Parliament. Usually the number of deputies per district depends on its population. The Netherlands is the extreme case. The entire country is a single 150-seat constituency. In a small, densely populated country like the Netherlands, an unique district may be appropriate. Spain is neither small nor densely populated. But, after 40 years of democracy we can say that deputies do not represent their constituencies, they represent their parties, they blindly follow their central party leaders.

The critics of proportional representation often point out that proportionality leads to instability. Nevertheless, the stability of a government is partly a function of its base in the electorate. Coalitions, which control a majority of seats in the legislature, are far more likely than single-party majority governments to represent a majority of the voters. It is a matter of balance between governability and representativeness.

## Elections in Spain

Spain is a parliamentary monarchy with two legislatures. These are the Congress of Deputies (_Congreso de los Diputados_) and the Senate (_Senado_). From 1977, a proportional representation electoral system was established for the lower chamber of parliament (Congress of Deputies) as well as for regional and local elections. Beside the provinces, Spain is divided in 17 autonomous communities which exercise legislative powers through their own parliamentary assemblies.

There are 350 seats in the Congress of Deputies. The electoral system uses closed party lists. Each one of Spain's fifty provinces is a constituency entitled to an initial minimum of two seats; the cities of Ceuta and Melilla elect one member each. The remaining 248 seats are allocated among the fifty provinces in proportion to their populations. This is constitutionalized.

The Senate is a chamber of territorial representation. 264 are chosen by direct election and 56 are selected through indirect vote by the parliaments of the autonomous communities. Every elector is able to vote three of the candidates on a single ballot.

The historical development of the electoral legislation in Spain after the Franco dictatorship can be found in Hopkin (2005):

> The current electoral arrangements in Spain were established in their essentials by the 1976 Law for Political Reform, and then by a decree-law passed in March 1977, little more than two months before the first democratic elections took place. These arrangements were in part incorporated into the 1978 constitution, and then confirmed almost in their entirety by an Electoral Law passed in 1985. Spain's current electoral rules are therefore a result of the peculiarities of the early stages of the transition process[...]  A conservative grouping in the Cortes, headed by former Franco minister Manuel Fraga, favoured a majoritarian system because of a (misplaced) belief that conservative forces would benefit[...] The opposition parties tended to favour proportional representation (PR), and Suárez's commitment to consensus ensured a compromise between PR and majoritarianism.

> [...] these restrictions also implied a significant over-representation of rural Spain, which Fraga's conservative faction deliberately engineered.

> The most important feature of the electoral arrangements for the Congress of Deputies is the choice of the province as the electoral constituency. Spain's fifty provinces were almost all established by the 1833 administrative reform[...] The provincial boundaries therefore reflect the distribution of the Spanish population in the first half of the nineteenth century, and take no account of the demographic upheavals resulting from the subsequent process of economic and social modernization[...] The minimum of two seats for each province, and the small size of the Congress of Deputies, produces a high degree of malapportionment.


This law (_Ley Orgánica del Régimen Electoral General, LOREG_) is based on d’Hondt system of proportional representation (largest average) for all the elections except for the Senate. It was conceived by the Belgian mathematician Victor d'Hondt in 1899. Nevertheless, in order to participate in the allocation of seats, a list must receive at least 3% of all valid votes cast in the constituency, including blank ballots. 

After a 40-year's dictatorship, this electoral system has been fruitful: unremarkably fragmented party system, stable governments, frequent absolute majorities in legislature; at the price of a huge disproportionality. As Colomer (2004) said:

> Disproportionality between party votes and seats has oscillated between 5 and 15 per cent, achieving the highest average values among proportional representation systems in Western Europe[...]

> As a whole, the electoral system used in Spanish democracy since 1977 has favoured the two larger Spain-wide parties as well as the large regional parties in the corresponding districts while hurting minor Spain-wide parties[...]

> The electoral system has also favoured the flourishing of multiple regional parties and the corresponding pressures in favour of political decentralization and strengthening of regional governments.

All electoral systems are imperfectly proportional. This is a mathematical truth (Balinski and Young, 2001). The translation of votes into seats is always distorted to some degree, usually in favour of the largest party. Constituencies with vastly different numbers of voters are a prime source of disproportional outcomes. So, why not to use the most proportional approach available? How to know when an electoral system is more proportional than other? What if legislatures in the past would have been selected by a single constituency without the lower bound of the 3% of all valid votes cast in the constituency? Would have single-party cabinets been so frequent? Would have there been any feasible coalition government? 

## Measures of disproportionality

A disproportionality index is an aggregated estimator of the difference between votes cast and seats allocation. There are at least 25 indexes to measure the disproportionality of an electoral system in the most cited literature: Monroe (1994), Taagepera and Grofman (2003), Grilli di Cortona et al. (1999) and Borooah (2002). A survey of some indexes can be found in Gallagher (1991), Aleskerov and Platonov (2005), Karpov (2008) and Bolun (2012). Each paper studies a different set of indexes, but all of them reach an equivalent conclusion: disproportionality indexes should be employed with great caution, basically because some voters will not vote for the preferred party if polls say they have no chance to gain a seat. Hence, none disproportionality index is a reliable proxy for the linkage among votes, seats and voters will. Probably for this reason, there is not a universally accepted index of disproportionality.

In the following, we assume that $n$ parties receive $V_1,V_2,\ldots,V_n$ votes in an election (in some articles $n$ represents the number of parties that have reached or exceeded the representation threshold). As a result, some seats in parliament are assigned to each party: $S_1,S_2,\ldots,S_n$. The percentages of votes ($v_i=\frac{V_i}{\sum V_j}(\%), i \in {1,\ldots,n}$) and the percentages of seats ($s_i=\frac{S_i}{\sum S_j} (\%), i \in {1,\ldots,n}$) will be used. In the ideal case, these ratios should be equal. All these indexes have $0$ as optimal value, unless otherwise stated.

* __The Rae index__: $ I_{Rae} = \frac{1}{n} \sum _{i=1}^n |v_i-s_i| $. It represents the mean of absolute deviations. It tends to have very low values, describing the system to be more proportional than it actually is, especially when $n$ is large. In a reviewed version, Rae disregarded parties with less than $0.5\%$ of the vote.
* __The Loosemore-Hanby index__: $ I_{LH} = \frac{1}{2} \sum _{i=1}^n |v_i-s_i| $. It measures the percentage of seats misallocated and tries to avoid the disadvantages of the Rae index. It is the most widely used index.
* __The Grofman index__: $ I_G = \frac{1}{N} \sum _{i=1}^n |v_i-s_i| $. $N$ is the effective number of parties: $N=\frac{1}{\sum {ratio_v}_i^2}$ (some papers use the alternative definition $N_s=\frac{1}{\sum {ratio_s}_i^2}$).
* __The Lijphart index__: $ I_L = \max\limits_{{1 \leq i \leq n}}{|v_i-s_i|} $. It shows the most inaccurately represented party. Some versions use the average of absolute deviations for the two largest parties.
* __The Gallagher index__: $ L_{sq} = \sqrt{\frac{1}{2} \sum _{i=1}^n (v_i-s_i)^2} $. It differs by a constant from the least square index. It weighs more strongly a large absolute deviation than lots of tiny ones. Lijphart (1994) characterizes this index as _the most faithful reflection of disproportionality of election results_. Maybe for that, it has recently become one of the most used.
* __The Saint-Lague index__: $ I_{SL} = \sum_{i=1}^n{\frac{1}{v_i} (v_i-s_i)^2} $. Gallagher (1991) considers this as the soundest index from a theoretical point of view.
* __The D'Hont index__: $ I_{DH} = \max{\frac{s_i}{v_i}} $. The optimal value is $1$.
* __The Cox-Shugart index__: $ I_{CS} = \frac{\sum_{i=1}^n v_i s_i}{\sum_{i=1}^n v_i^2} $. It is the slope of the regression line of  $s_i$ over $v_i$ by least squares. The optimal value is $1$. When largest parties are overrepresented  $b > 1$. Other versions add a bias term to the regression line.
 
It is hard to compare the indexes directly. They have different meanings and use different units (check the reference section for some benchmarks). Nevertheless, we will plot their values over time in a certain kind of election (here legislative Spanish elections using 52 constituencies and 3% threshold). Moreover, we will calculate an alternative apportionment by introducing some minor changes in the electoral law for each election (a single constituency with no threshold). For this simulated seat distributions the very same indexes will be estimated and plotted. From these plots, we might deduce how disproportionality has evolved in General Elections in Spain and if those minor changes could have improved the representativeness of the people's will in Parliament.

## Measures of multipartism

The simplest indicators for multipartism found in the literature are:

- The total number of parties contesting an election.
- The total number elected to the legislature.
- The vote and seat shares of the largest party.
- The vote and seat shares of the two largest parties combined.

As it was explained in the previous section, the Grofman index of disproportionality uses the effective number of parties. This indicator can be defined in two different ways:

$$ N = \frac{1}{\sum {ratio_v}_i^2} \mbox{ or }  N = \frac{1}{\sum {ratio_s}_i^2} $$

This index was introduced in Laakson and Taagepera (1979) and shows how many parties there are in Parliament, taking their relative sizes into account.  For instance, in a legislature with two equally strong parties, the effective number of parties is exactly $2.0$; for three equal parties it is $3.0$; for two strong parties and one weaker party, it will be somewhere in the interval $[2,3]$. Lijphart (2012) recommends the version using the ratio of seats because it is a proxy of the strengths and patterns of parties in parliaments and on their effects on the formation of cabinets. Dunleavy and Boucek (2003) show that $N$ does not deliver a reliable or consistent relationship between changes in the largest party’s share of the vote and the index number shown.

Here, the number of parties with a seat in Parliament and the two version of the effective number of parties will be plotted over time for the two scenarios described in the previous section.

## Measures of governability

There is a trade-off between representativeness and governability. Proportional systems try to maximize representativeness in exchange of difficult governance and legislative fragmentation. Probably that is the reason why few political scientists would advocate a pure proportional system —with virtually no minimum threshold of the vote necessary for entry into parliament— that existed for some time in Italy and Israel.

It is hard to find an index to quantify the governability of an election outcome. It depends on the local political culture, legislation, parties and history. In order to check if the proposed two minor change in the Spanish electoral law would have leaded to a lack of governability, parties will be grouped by their ideology. Five groups will be considered:

- Left-wing parties include social democrats, socialists, communists, and revolutionaries.
- Center-right parties include conservatives, reformists, demochristians, and Francoists.
- Nationalist parties include regionalist and moderate nationalists. They can reach agreements with major parties within the two previous categories.
- Soberanist or pro-independence parties. Some parties from the previous category can be included here and some from here can be included there.
- Ecologist parties.

It is assumed that parties within a group are likely to form coalitions in Parliament and/or cabinet. Thus, when one of these groups achieve an overall majority, that Parliament and the cabinet produced will be considered stable.

## Disproportionality indexes over time

Each plot represents one disproportionality index and contains two time series: in blue the values for the actual electoral law and in orange the values produced by a apportionment for a single district with no electoral threshold.

Two main conclusions from the plots:

- there is a trend to reduce disproportionality over time;
- as expected, the orange scenario is always much more proportional than the blue one.

```python
sc_selector = indicators_ts_df['Single_Constituency']

fig, axlist = plt.subplots(4, 2)
fig.set_size_inches(10, 20)
indicator_names = ['Rae', 'Loosemore-Hanby', 'Gallagher', 'Grofman', 
                   'Lijphart', 'Saint-Lague', 'DHont', 'Cox-Shugart']
indicators = [x.lower().replace('-','_') for x in indicator_names]
plt.grid(True)

for idx, ax in enumerate(axlist.flatten()):
    indicator = indicators[idx]
    name = indicator_names[idx]
    optimal = 1 if name in ['DHont', 'Cox-Shugart'] else 0
    indicators_ts_df[~sc_selector][indicator].plot(ax=ax, title='%s Index' % name, 
                                                   label='52 Constituencies')
    indicators_ts_df[sc_selector][indicator].plot(ax=ax, label='Single Constituency')
    ax.grid(True)
    ax.axhline(y=optimal, linestyle='--', color='g', label='Ideally Proportional')

fig.subplots_adjust(top=0.9, bottom=0.15, left=0.1, right=0.95, hspace=0.25, wspace=0.25)
leg = axlist.flatten()[0].legend(loc='upper right', bbox_to_anchor=(1.4, 1.3))
```

## Multipartism indexes over time

In theory, a pure proportional electoral system as the one proposed would lead to a greater fragmented parliament. Nevertheless, there is a case in which this frequent fact is not accomplished. As the plot of the effective number of parties by seats reveals, the difference between the two scenarios proposed is at most one unit. This is a reasonable price in order to gain much more representativeness.

```python
fig, axlist = plt.subplots(3, 1)
fig.set_size_inches(7, 14)
titles = ['#parties in Parliament', 'Effective # of parties by votes', 
          'Effective # of parties by seats']
columns = [SEATS, EFFECTIVE_NUMBER_OF_PARTIES_BY_VOTES, EFFECTIVE_NUMBER_OF_PARTIES_BY_SEATS]
plt.grid(True)

for idx, ax in enumerate(axlist.flatten()):
    
    dataframe = seats_ts_df if idx == 0 else eff_n_parties_ts_df
    sc_selector = dataframe[SINGLE_CONSTITUENCY]
    dataframe[~sc_selector][columns[idx]].plot(ax=ax, title=titles[idx], 
                                               label='52 Constituencies')
    dataframe[sc_selector][columns[idx]].plot(ax=ax, title=titles[idx], 
                                              label='Single Constituency')
    ax.grid(True)

fig.subplots_adjust(top=0.9, bottom=0.15, left=0.1, right=0.95, hspace=0.25, wspace=0.25)
leg = axlist.flatten()[0].legend(loc='upper center', bbox_to_anchor=(0.5, 1.3))
```

## Conclusions

Conventionally, it is thought that majoritarian elections promote a party system with relatively few parties (typically two), which amalgamating different political tendencies. As a consequence, it produces durable governments supported by wide majorities in legislature. On the contrary, list-based proportional representation is said to facilitate the representation of social cleavages, leading to fragmented legislature and unstable cabinets. Spanish politics is described by the former, but legally regulated by the latter. As Hopkin (2005) titled one of his articles: _Spain: Proportional Representation with Majoritarian Outcomes_.

It is also thought that the D'Hont formula is the least proportional among the highest averages methods, and systematically favors the larger parties. But this is not a natural law of physics, as 2017 Catalan election showed. If the two greatest pro-independence parties had joined in an electoral coalition, the apportionment would have been exactly the same. (See [Sistemas Electorales (in Spanish)](./Sistemas Electorales.ipynb)). Hence, the electoral formula is not the reason why disproportional parliaments are produced by the Spanish law.

The district magnitude and the electoral threshold are more explanatory variables for the disproportionality of the Spanish electoral system. Removing both by setting a single nationwide district and not requiring any minimal percentage of votes to enter in the apportionment is enough to achieve a much more proportional electoral system. Using a single district requires  a constitutional amendment, and consequently a qualified majority of three-fifths in both Chambers.

Two political reasons are used to avoid these changes in the law. On one hand, districts or territories should be represented in the Parliament. Nevertheless, deputies in the Congress represent their parties, not their districts, as any Spaniard perfectly knows after 40 years of democracy. In addition, any region has their own autonomous Parliament and the Senate is a territorial chamber. Thus, it does not seem reasonable to assign seats in the Congress by region.

On the other hand, not requiring a minimal threshold would increase enormously the number of parties with a seat, making impossible to govern or legislate. This could be true in the first years of the Spanish democracy after the dictatorship, but it doesn't seem to be true anymore. The bias in favour of the right wing and against the middle-size and ecologist nationwide parties is not just unfair, but also counter-productive to ease governability. The outcome has been weak cabinets supported by moderate nationalist parties, because minorities, that are not geographically concentrated, are under-represented in Parliament.

Pre-Mussolini Italy, France’s Fourth Republic, and the Weimar Republic are the usual examples against proportional representation (Quade (1999)). But as Lijphart (1991) says _disproportional election results are inherently unfair and undemocratic...any electoral system that does not ensure the fair and full representation of each group risks the group's alienation and rejection of the system_. This warning is particularly important in the Spanish case. A wrong and unfair electoral system may produce undesired political consequences:

> The choice of a majoritarian electoral system by the leaders of the Spanish Second Republic established in 1931 can explain the high and increasing levels of political polarization which eventually led to revolution, _coup d’état_ and ultimately civil war[...]

> In contrast, the following, much more successful, democratic experience in Spain, which started in 1977, has been based on elections by proportional representation[...] Approving an electoral system able to prevent some of the worse
consequences of the one implemented in the 1930s was also a concern of most political actors in the 1970s.

> The outburst of a polarized conflict at the induction of electoral rules rather than from social unrest greatly surprised many Spaniards. The lack of understanding why civil war had occurred fed a great fear that a similar conflict might reappear in the future. Colomer (2004).
